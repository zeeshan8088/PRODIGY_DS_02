```python
# -------------------------------
# 1. Importing Libraries
# -------------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
sns.set_palette('pastel')

# -------------------------------
# 2. Load the Dataset
# -------------------------------
df = pd.read_csv('train.csv')  # Make sure train.csv is uploaded to your environment
print("Data Loaded!")
print(df.head())

# -------------------------------
# 3. Dataset Overview
# -------------------------------
print("\nShape of dataset:", df.shape)
print("\nColumn Summary:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())

# -------------------------------
# 4. Data Cleaning
# -------------------------------

# Fill missing Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Create 'HasCabin' from missing Cabin values
df['HasCabin'] = df['Cabin'].notnull().astype(int)

# Drop Cabin (too many missing values)
df.drop('Cabin', axis=1, inplace=True)

# Drop Ticket (not useful for EDA)
df.drop('Ticket', axis=1, inplace=True)

# Convert 'Sex' and 'Embarked' to numeric for analysis
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

print("\nData cleaned and ready for analysis!")

# -------------------------------
# 5. Exploratory Data Analysis (EDA)
# -------------------------------

# A. Survival Count
sns.countplot(x='Survived', data=df)
plt.title("Counts of Survival (0 = No, 1 = Yes)")
plt.show()

# B. Survival Rate by Sex
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Sex (0=Male, 1=Female)")
plt.show()

# C. Survival Rate by Passenger Class
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Ticket Class (1=1st, 3=3rd)")
plt.show()

# D. Age Distribution by Survival
sns.histplot(data=df, x='Age', bins=30, hue='Survived', multiple='stack')
plt.title("Age Distribution by Survival")
plt.xlabel("Passenger Age")
plt.ylabel("Count")
plt.show()

# E. Fare Distribution by Class
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare Amount by Ticket Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare ($)")
plt.show()

# -------------------------------
# 6. Correlation Heatmap (Fixed)
# -------------------------------
# Only use numeric columns (drop non-informative if needed)
numeric_df = df.select_dtypes(include=['number'])
if 'PassengerId' in numeric_df.columns:
    numeric_df = numeric_df.drop(['PassengerId'], axis=1)

plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()

# -------------------------------
# 7. Insights & Observations
# -------------------------------
print("\n Key Insights:")
print("- Female passengers had a much higher survival rate.")
print("- Passengers in 1st class had better chances of surviving.")
print("- Younger children also had slightly greater survival rate.")
print("- Passengers who paid higher fares were more likely in 1st class.")
```
